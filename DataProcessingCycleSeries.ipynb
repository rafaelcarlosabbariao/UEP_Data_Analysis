```python
df = pd.read_csv("D:/Python/Machine Learning/Research/30%Success_UEP_C071018A_084_NonSlit_A_30DoD_072418_091718-01-103 (1).csv")
df.head()
df.info()
```

```python
X = df.iloc[:, 0].values # [row, columns]
y_1 = df.iloc[:, [1]].values # charge capacity
y_2 = df.iloc[:, [2]].values # discharge capacity
# 1, 2 --> charge, discharge capacity (Ah)
# 6, 7 --> charge, discharge energy (Wh)
# 14, 15 --> minimum, maximum potential (V)
# 31, 32 --> maximum charge, minimum discharge power (W)

y1 = plt.plot(X, y_1, color = 'red', label = 'Charge Capacity')
y2 = plt.plot(X, y_2, color = 'blue', label = 'Discharge Capacity')
plt.xlabel('Cycle Number')
plt.ylabel('Charge/Discharge Capacity (Ah)')
plt.legend(loc='lower left')
```
